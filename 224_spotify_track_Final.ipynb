```python
from google.colab import drive
import pandas as pd
import os # Connects this Colab notebook to your Google Drive
drive.mount('/content/drive')
# Set the path to the folder containing the CSV file
folder_path = '/content/drive/MyDrive/ITE224_FINAL_PROJECT'
file_path = os.path.join(folder_path, 'dataset.csv')
```

```python
# Load the CSV file into a pandas DataFrame named 'df’
df = pd.read_csv(file_path)
df.head(6) # Display the first 6 rows to get a quick look at the data
```

```python
df = df.drop(columns=['Unnamed: 0'], errors='ignore')
df.shape
```

```python
# Show information about the columns and data types
df.info()
```

```python
# Show summary statistics
df.describe()
```

```python
# Step 2: Data Cleaning & Preprocessing

# Convert 'explicit' (TRUE/FALSE) to numeric 1/0
df['explicit'] = df['explicit'].astype(str).str.upper().map({'TRUE': 1, 'FALSE': 0})

# Make loudness positive (since dB are negative)
df['loudness_pos'] = -df['loudness']

# Create binary target (1 = popular song, 0 = not)
df['hit'] = (df['popularity'] >= 70).astype(int)

# Drop text/id columns not useful for model
df = df.drop(columns=['track_id', 'artists', 'album_name', 'track_name'], errors='ignore')

# One-hot encode genre if exists
if 'track_genre' in df.columns:
    df = pd.get_dummies(df, columns=['track_genre'], drop_first=True)

# Fill missing numeric values with column median
df = df.fillna(df.median(numeric_only=True))

# Show first few rows to verify
df.head()

```

```python
#Step 2 (EDA): Visualizations

import matplotlib.pyplot as plt  # Import matplotlib for plotting graphs

# --- Histogram ---
# Plot a histogram to see the distribution of the 'popularity' column
df['popularity'].plot(
    kind='hist',         # Histogram type plot
    bins=30,             # Divide data into 30 bins
    title='Popularity Distribution'  # Title of the plot
)
plt.xlabel('popularity')  # Label the x-axis
plt.show()                # Display the histogram

# ---- Box plot ---
# Create a boxplot to compare 'danceability' values for hit vs non-hit songs
df.boxplot(
    column='danceability',  # Column to plot
    by='hit'                # Group data by the 'hit' column
)
plt.title('Danceability by Hit')  # Title of the plot
plt.suptitle('')                  # Remove the automatic extra title matplotlib adds
plt.show()                        # Display the box plot



```

```python
#  Step 3: Split & Train

from sklearn.model_selection import train_test_split   # For splitting dataset into train & test sets
from sklearn.preprocessing import StandardScaler       # For scaling/normalizing features
from sklearn.linear_model import LogisticRegression    # Logistic Regression model
from sklearn.tree import DecisionTreeClassifier        # Decision Tree model
import numpy as np # Added numpy import

#  Remove both 'hit' (target) and 'popularity' (to avoid data leakage) from the features
X = df.drop(columns=['hit', 'popularity'])  # X = all columns except target & unwanted feature
y = df['hit']                               # y = target column (hit or not)

# Keep only numeric columns (drop text/categorical columns)
X = X.select_dtypes(include=[np.number])

# Scale features for Logistic Regression (important because it’s sensitive to feature scale)
scaler = StandardScaler()       # Create scaler
X_scaled = scaler.fit_transform(X)  # Fit on X and transform to scaled values (mean=0, std=1)

# Split dataset into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,                  # Features after scaling
    y,                         # Target values
    test_size=0.2,             # 20% of data for testing
    random_state=42,           # Reproducible split
    stratify=y                 # Keep class ratio same in train/test
)

# Train Logistic Regression model
log_reg = LogisticRegression(max_iter=1000, solver='liblinear')  # max_iter avoids convergence warnings
log_reg.fit(X_train, y_train)  # Fit model to training data

# Train Decision Tree model
tree = DecisionTreeClassifier(random_state=42)  # Random state for reproducibility
tree.fit(X_train, y_train)                      # Fit model to training data
```

```python
# Step 4: Evaluate & Compare

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report  # Metrics for evaluation
import numpy as np                                   # Numerical utilities (optional here)
import seaborn as sns                                # For pretty confusion-matrix heatmaps
import matplotlib.pyplot as plt                      # Plotting library

#  Predictions
y_pred_lr = log_reg.predict(X_test)                  # Predicted labels from Logistic Regression
y_pred_tree = tree.predict(X_test)                   # Predicted labels from Decision Tree

# Accuracy
acc_lr = accuracy_score(y_test, y_pred_lr)           # Accuracy of Logistic Regression
acc_tree = accuracy_score(y_test, y_pred_tree)       # Accuracy of Decision Tree

print(f"Logistic Regression Accuracy: {acc_lr:.3f}") # Print LR accuracy to 3 decimals
print(f"Decision Tree Accuracy: {acc_tree:.3f}")     # Print Tree accuracy to 3 decimals

# Classification reports
print("\nLogistic Regression Report:")               # Header for LR metrics
print(classification_report(y_test, y_pred_lr))      # Precision/Recall/F1/Support for LR

print("\nDecision Tree Report:")                     # Header for Tree metrics
print(classification_report(y_test, y_pred_tree))    # Precision/Recall/F1/Support for Tree

#  Confusion Matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 5))      # Create a 1x2 subplot grid

# Heatmap for Logistic Regression confusion matrix
sns.heatmap(
    confusion_matrix(y_test, y_pred_lr),             # Confusion matrix array (true vs predicted)
    annot=True,                                      # Show counts in cells
    fmt='d',                                         # Integer formatting
    ax=axes[0],                                      # Plot on the left axis
    cmap='Blues'                                     # Color map
)
axes[0].set_title('Logistic Regression Confusion')   # Title
axes[0].set_xlabel('Predicted')                      # X-axis label
axes[0].set_ylabel('True')                           # Y-axis label

# Heatmap for Decision Tree confusion matrix
sns.heatmap(
    confusion_matrix(y_test, y_pred_tree),           # Confusion matrix for Tree
    annot=True,
    fmt='d',
    ax=axes[1],                                      # Plot on the right axis
    cmap='Greens'
)
axes[1].set_title('Decision Tree Confusion')         # Title
axes[1].set_xlabel('Predicted')                      # X-axis label
axes[1].set_ylabel('True')                           # Y-axis label

plt.show()                                           # Render both heatmaps

#  Accuracy comparison bar chart
plt.bar(['Logistic Regression', 'Decision Tree'],    # Bar labels
        [acc_lr, acc_tree])                          # Bar heights (accuracies)
plt.ylabel('Accuracy')                               # Y-axis label
plt.title('Model Accuracy Comparison')               # Plot title
plt.ylim(0, 1)                                       # Keep y-axis in [0,1]
for i, v in enumerate([acc_lr, acc_tree]):           # Add text labels above bars
    plt.text(i, v + 0.01, f"{v:.2f}", ha='center')   # Position text slightly above each bar
plt.show()                                           # Display bar chart

```

```python
# Step 5: Feature Importance & Conclusion

import numpy as np                          # For numerical operations (argsort, abs)
import matplotlib.pyplot as plt              # For plotting
from sklearn.metrics import roc_curve, auc   # For ROC curve and AUC

#  (Viz 1) Decision Tree: Feature Importance
imp = tree.feature_importances_              # Importance scores learned by the Decision Tree
idx = np.argsort(imp)[::-1]                  # Indices sorted by importance (descending)
names = X.columns[idx]                       # Feature names in the same sorted order

top_k = min(15, X.shape[1])                  # Plot up to 15 (or fewer if <15 features)
plt.figure(figsize=(10, 5))                  # Set figure size
plt.bar(range(top_k), imp[idx][:top_k])      # Bar chart of top-k importances
plt.xticks(range(top_k), names[:top_k], rotation=90)  # Feature labels (rotated for readability)
plt.title('Decision Tree Feature Importance (Top {})'.format(top_k))  # Title
plt.ylabel('Importance')                     # Y-axis label
plt.tight_layout()                           # Prevent label cutoff
plt.show()                                   # Display the plot

#  (Viz 2) Logistic Regression: Coefficient Importance
# Note: We trained LR on scaled features, so coefficients are comparable across features.
coef = log_reg.coef_[0]                      # Coefficients for each feature (shape: [n_features])
coef_abs = np.abs(coef)                      # Use absolute value to measure "strength" regardless of sign
idx_lr = np.argsort(coef_abs)[::-1]          # Sort by absolute magnitude (descending)
names_lr = X.columns[idx_lr]                 # Reorder feature names accordingly

top_k_lr = min(15, X.shape[1])               # Plot top 15 (or fewer)
plt.figure(figsize=(10, 5))                  # Set figure size
plt.bar(range(top_k_lr), coef_abs[idx_lr][:top_k_lr])  # Bar chart of |coefficients|
plt.xticks(range(top_k_lr), names_lr[:top_k_lr], rotation=90)  # Feature labels
plt.title('Logistic Regression |Coefficient| (Top {})'.format(top_k_lr))  # Title
plt.ylabel('|Coefficient| (on scaled features)')  # Y-axis label
plt.tight_layout()                           # Prevent label cutoff
plt.show()                                   # Display the plot

```
